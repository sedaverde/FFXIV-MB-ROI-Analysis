```python
## Dependencies
import csv
import pandas as pd
import numpy as np
```

```python
## Open recipe.csv from xivapi/ffxiv datamining repo; this is the information for for all craftable items in ffxiv
recipe = pd.read_csv ('https://raw.githubusercontent.com/xivapi/ffxiv-datamining/master/csv/Recipe.csv',header = 1)
recipe
```

```python
## Drop row that identified the data types for each column
recipe = recipe.drop(labels=0, axis=0)
recipe
```

```python
## Drop columns that are not needed
recipe = recipe.drop(recipe.columns[[26,32,38,41,46]],axis = 1)
```
