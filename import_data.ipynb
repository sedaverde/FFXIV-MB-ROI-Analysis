```python
## Dependencies
import csv
import pandas as pd
import numpy as np
```

```python
## Open recipe.csv from xivapi/ffxiv datamining repo; this is the information for all craftable items in ffxiv
recipe = pd.read_csv ('https://raw.githubusercontent.com/xivapi/ffxiv-datamining/master/csv/Recipe.csv',header = 1)
recipe
```

```python
## Drop row that identified the data types for each column
recipe = recipe.drop(labels=0, axis=0)
recipe
```

```python
## Drop columns that are not needed
recipe = recipe.drop(recipe.columns[[26,28,29,31,32,38,41,46]],axis = 1)
```

```python
## Verify the data types that the data was imported into as, in order to identify those needing updating
recipe.dtypes
```

```python
## Simplify/clarify column names; ItemID (DBkey)
recipe =recipe.rename(columns={'Number':'RecipeID','RecipeLevelTable':'Level','Item{Result}':'ItemID','Amount{Result}':'Quant','Item{Ingredient}[0]':'Mat0','Item{Ingredient}[1]':'Mat1','Item{Ingredient}[2]':'Mat2','Item{Ingredient}[3]':'Mat3','Item{Ingredient}[4]':'Mat4','Item{Ingredient}[5]':'Mat5','Item{Ingredient}[6]':'Mat6','Item{Ingredient}[7]':'Mat7','Item{Ingredient}[8]':'Mat8','Item{Ingredient}[9]':'Mat9','Amount{Ingredient}[0]':'Quant0','Amount{Ingredient}[1]':'Quant1','Amount{Ingredient}[2]':'Quant2','Amount{Ingredient}[3]':'Quant3','Amount{Ingredient}[4]':'Quant4','Amount{Ingredient}[5]':'Quant5','Amount{Ingredient}[6]':'Quant6','Amount{Ingredient}[7]':'Quant7','Amount{Ingredient}[8]':'Quant8','Amount{Ingredient}[9]':'Quant9','RequiredCraftsmanship':'Craftsmanship','RequiredControl':'Control','QuickSynthCraftsmanship':'QSCraftsmanship','QuickSynthControl':'QSControl','SecretRecipeBook':'Book','CanQuickSynth':'QS','CanHq':'HQ','IsSpecializationRequired':'Specialized','IsExpert':'Expert'})
```

```python
## Set correct data types
recipe = recipe.astype({'RecipeID':'int','CraftType':'byte','Level':'byte','ItemID':'int','Quant':'byte','Mat0':'int','Mat1':'int','Mat2':'int','Mat3':'int','Mat4':'int','Mat5':'int','Mat6':'int','Mat7':'int','Mat8':'int','Mat9':'int','Quant0':'byte','Quant1':'byte','Quant2':'byte','Quant3':'byte','Quant4':'byte','Quant5':'byte','Quant6':'byte','Quant7':'byte','Quant8':'byte','Quant9':'byte','Craftsmanship':'uint16','Control':'uint16','QSCraftsmanship':'uint16','QSControl':'uint16','Book':'byte','QualityFactor':'int','Status{Required}':'int','Item{Required}':'int'})
```

```python
## Convert object to Boolean
recipe = recipe.replace({'True': True, 'False': False})
```

```python
## Open item.csv from xivapi/ffxiv datamining repo; this is the information for all items in ffxiv; only importing certain columns
item = pd.read_csv("https://raw.githubusercontent.com/xivapi/ffxiv-datamining/master/csv/Item.csv",header = 1,usecols = ['#','Description','Name','StackSize','Price{Mid}','Price{Low}','CanBeHq','IsDyeable','ClassJob{Use}','IsUntradable'])
item
```

```python
## Drop row that identified the data types for each column
item = item.drop(labels=0, axis=0)
## Dropping items that begin with the word Dated and items that are not tradable
item = item[item['Name'].str.contains("Dated ")==False]
item = item[item['IsUntradable'].str.contains("True")==False]
## Rename # column to ItemID (DB key)
item = item.rename(columns={'#':'ItemID','Price{Mid}':'PriceMid','Price{Low}':'PriceLow','ClassJob{Use}':'ClassJobUse'})
item
```

```python
## Set correct data types
item = item.astype({'ItemID':'int','StackSize':'int','Description':'string','Name':'string','PriceLow':'int','PriceMid':'int','ClassJobUse':'int'})
item = item.replace({'True': True, 'False': False})
```
